```python
#  creating  2nd   dataframe
raw_data={
                 'Hour':['11','11','11','11','11'],
                 'Requests':['171','342','203','213','49'],
                 'Completes':['151','298','164','155','44']}
df_b=pd.DataFrame(raw_data,columns=['Hour','Requests','Completes'])
df_b

```

USING   THE  CONCAT  FUNCTION  TO DO CONCATENATION

```python
df_new=pd.concat([df_a,df_b], ignore_index=True)
df_new

# so the  rows  are  concatenated 
```

```python

# Creating  datafraame  to  perform a  Join  operation (inner -join)
raw_data={
              'Customers_name':['Rahul verma','Anurag Kumar','Sakshi mehta','ansh kumar','Abhi bhatt','Ritu kumari','Prachi desai','Saurav k','B Murli','Poonam dave'],
              'Hour':['11','11','11','11','11','11','11','11','11','11'] }
df_n =pd.DataFrame(raw_data, columns=['Customers_name','Hour'])
df_n
```

NOW  DOING  A INNER   TO  THE  ABOVE  DATA

```python
pd.merge(df_new,df_n ,on='Hour')
```
