```python
# storing   the  above  9 cols  in  a  diffrent  variable  named Fdi_data1
Fdi_data1 = Fdi_data.iloc[:,0:9]
```

```python
Fdi_data1.describe()
```

```python
from matplotlib import pyplot
from pandas import read_csv
names = ['2000-01','2001-02','2002-03','2004-05','2005-06','2006-07','2007-08']
data = (Fdi_data1, names=names)
correlations = data.corr()
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
```
