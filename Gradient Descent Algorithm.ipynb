#### Stochastic Gradient Descent

```python
# importing  the libaries 
from sklearn import linear_model 
from sklearn.datasets import load_boston
from sklearn.linear_model import SGDRegressor

from sklearn.decomposition import PCA
from sklearn.preprocessing  import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.metrics  import r2_score,mean_squared_error
from sklearn.pipeline  import Pipeline

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
# now  splitting  the  data 
x = load_boston()
#crating  the  Dataframe
data = pd.DataFrame(x.data, columns=x.feature_names)
y=x.target
y =pd.DataFrame(y)


#now  creating  our  model using  sklearn pipeline
pipe=[]
pipe.append(('SC',StandardScaler()))
pipe.append(('PCA',PCA(n_components=8)))
pipe.append(('SGDR',SGDRegressor(alpha=0.1,learning_rate='optimal', max_iter=400, penalty=12)))
model=Pipeline(pipe)
model

```

```python
#cross_validation_score
cv_results= cross_val_score(model, data, y, cv=5)

msg = '%s: %f (%f)'% ('SGDRegressor',cv_results.mean(), cv_results.std())
print(msg)
```
