```python
df = df.drop(['full_name', 'b_day', 'height', 'weight', 'college'], axis = 1)
df.head()
```

```python
def removehash(value):
    value = value[1:]
    return int(value)
```

```python
df['jersey'] = df['jersey'].apply(removehash) #  we  removed  the  hash from  the  col
```

```python
df.head()
```

```python
df['salary'] = df['salary'].apply(removehash)
df.head()   #  removed   hash  from the  salary also
```

```python
df['team'].value_counts()
```

```python
df['country'].value_counts()
```

```python
def removecountryoutlier(value):
    if value not in('USA','Canada','Australia'):
        return 'others'
    else:
        return value
df['country']= df['country'].apply(removecountryoutlier)
df['country'].value_counts()
    
```

```python
df['position'].value_counts()
```

```python
def removeundrafted(value):
    if value=='Undrafted':
        return 
    else:
        return value
    
df['draft_round'] = df['draft_round'].apply(removeundrafted)
df['draft_peak'] = df['draft_peak'].apply(removeundrafted)

df = df.dropna()
```

```python
df.isnull().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['position'] = le.fit_transform(df['position'])
df['country'] = le.fit_transform(df['country'])
df['team'] = le.fit_transform(df['team'])
# starting   with  encoding   the  categorical  columns 
```
